```python
from casadi import MX, vertcat, nlpsol

# Decision variables
x = MX.sym('x')  # Pro tablets
y = MX.sym('y')  # Mini tablets

# Objective: Maximize 182 * x + 139 * y â†’ Minimize -182 * x - 139 * y
profit = -182 * x - 139 * y

# Inequality constraints (Ax â‰¤ b)
g1 = 6 * x + 9 * y   # Labor
g2 = x + y           # Chip sets
g3 = 15 * x + 10 * y # Electronic components

# Variable vector
vars = vertcat(x, y)

# Define problem
nlp = {
    'x': vars,
    'f': profit,
    'g': vertcat(g1, g2, g3)
}

# Bounds for decision variables: x â‰¥ 0, y â‰¥ 0
lbx = [0, 0]
ubx = [1e5, 1e5]  # Large number as upper bound (no constraint)

# Inequality constraints: g â‰¤ b â†’ define as upper bounds only
lbg = [-1e20, -1e20, -1e20]  # No lower bounds
ubg = [7000, 1000, 14000]

# Solve with IPOPT
solver = nlpsol('solver', 'ipopt', nlp)
sol = solver(lbx=lbx, ubx=ubx, lbg=lbg, ubg=ubg)

# Extract solution
x_opt = float(sol['x'][0])
y_opt = float(sol['x'][1])
profit_max = -(float(sol['f']))  # Negate because we minimized -profit

# Output
print("Optimal solution found!")
print(f"Number of Pro tablets to produce: {x_opt:.2f}")
print(f"Number of Mini tablets to produce: {y_opt:.2f}")
print(f"Maximum profit: ${profit_max:.2f}")
```
